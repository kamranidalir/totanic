```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import kurtosis
from scipy.stats import skew

# ایجاد داده‌ها
data = np.random.normal(100, 15, 1000)
df =pd.DataFrame(data)

# رسم نمودار هیستوگرام
# plt.hist(data)
# plt.show()

# QQ_PLOT
# stats.probplot(data, dist="norm", plot=plt)
# plt.title("QQ")
# plt.show()

# sns
sns.displot(data, kind="kde")

kurt = kurtosis(data)
print(kurt)

if kurt > 0:
    print("نمودار leptokurtic (heavy-tailed) است.")
elif kurt < 0:
    print("نمودار platykurtic (light-tailed) است.")
else:
    print("نمودار دارای کرتوزیس نرمال است.")


skewness_value = skew(data)
print(f"مقدار چولگی: {skewness_value:.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df=pd.read_csv('/content/train_Titanic.csv')
# df.head()
# df.info()
# df.shape
# df.head(3)
def f(name):
    if '.' in name:
        return name.split(',')[1].split('.')[0].strip()
    else:
        return 'Unknown'
def g(tt):
    if tt in ['Mr']:
        return 1
    elif tt in ['Master']:
        return 3
    elif tt in ['Ms', 'Mlle', 'Miss']:
        return 4
    elif tt in ['Mrs','Mme']:
        return 5
    else:
        return 2

df['title'] = df['Name'].apply(f).apply(g)
df.head(5)

t = pd.crosstab(df['title'], df['Survived'])
t

sns.heatmap(t, annot=True, cmap='Blues', fmt='g')
plt.title('Survival Status by Title')
plt.show()


df = df.drop(['PassengerId','Name','Ticket'],axis=1)
df.head()

edt  = pd.get_dummies(df['Embarked'] )
edt.replace({True: 1, False: 0},inplace=True)


edt.drop(['S'], axis=1, inplace=True)
df = df.join(edt)
df

df.drop(['Embarked'], axis=1,inplace=True)
df.head(3)

df['Age'] = df['Age'].fillna(df['Age'].mean())
df.info()
sur=df["Survived"].value_counts()
# print(df.head(3))
# print(sur)

#/////////////////////////////////////////////////////////////////////////////////////////////////////////

# df1 = pd.DataFrame({'Bird' : ['A', 'A', 'B', 'B', 'B'],'Speed' : [380, 370, 24, 26,np.nan]})
# df1

# df1.groupby(['Bird']).mean()

# df1['Speed'] = df1.groupby(['Bird'])['Speed'].transform(lambda x: x.fillna(x.mean()))
# df1

# s = 9
# y = lambda a: a + 1
# print(y(s))

#/////////////////////////////////////////////////////////////////////////////////////////////////////////

df['Age'] = df.groupby(['Pclass'])['Age'].transform(lambda x: x.fillna(x.mean()))


df.drop("Cabin",axis=1,inplace=True)
df['Sex'] = df['Sex'].replace({'female': 0, 'male': 1})
df.head(10)
#تعداد مرده و زنده اقا و خانم
df_grouped = df.groupby(['Sex', 'Survived'])['Survived'].count().unstack()
print(df_grouped)



correlation = df['Sex'].corr(df['Survived'])
print("ضریب همبستگی پیرسون:", correlation)

# محاسبه ضریب همبستگی رتبه ای اسپیرمن (در صورت وجود داده های ترتیبی)
spearman_correlation = df['Sex'].corr(df['Survived'], method='spearman')
print("ضریب همبستگی رتبه ای اسپیرمن:", spearman_correlation)


# رسم هیستوگرام برای سن
plt.hist(df['Age'])
plt.xlabel('age')
plt.ylabel("people count")
plt.title('age histo')
plt.show()

# رسم هیستوگرام مجزا برای نجات یافتگان و فوت شدگان
plt.hist(df[df['Survived'] == 1]['Age'], label='survived', alpha=0.5)
plt.hist(df[df['Survived'] == 0]['Age'], label='drowns', alpha=0.5)
plt.xlabel('age')
plt.ylabel('people count')
plt.title("age histo")
plt.legend()
plt.show()
# s = sorted(df['Sex'].unique())
# s

# gm = dict(zip(s, range(0, len(s) + 1)))
# gm

# df['Sex'].head(5)

# df['Sex'] = df['Sex'].map(gm).astype(int)
# df['Sex'].head(5)
# سن کمترین و مسن ترین را پیدا کنید

df['Age'] = df['Age'].astype(int)
df.head(10)
min_age = df['Age'].min()
max_age = df['Age'].max()

# نتایج را چاپ کنید
print(f"کمترین سن: {min_age}")
print(f"مسن ترین سن: {max_age}")
#df["Age"].isna().any()
# print(df.head(800))
min_age_by_sex = df.groupby('Sex')['Age'].min()

# Print the minimum age for each sex
print("Minimum age by sex:")
print(min_age_by_sex)


row_numbers_of_min_age_individuals = df[df['Age'].isin(min_age_by_sex)].index

# Print the row numbers
print("Row numbers of individuals with the minimum age:")
print(row_numbers_of_min_age_individuals)

print(df.to_string())
```
